```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.common.exceptions import TimeoutException , WebDriverException
from time import sleep,ctime
from selenium.webdriver.common.by import By
```

## Extracting websites from the original dataset

```python
df = pd.read_csv('top-1m.csv' , names=['rank','website'])

site_df = df.iloc[700:1000]

print(site_df)
```

## Adding two extra columns to store results

```python
site_df.insert(2, "raw_dns", int(0)) 
site_df.insert(3, "final_dns", int(0)) 
print(site_df)
```

```python
firefox_domains = ["ocsp.digicert.com" , "shavar.services.mozilla.com", "push.services.mozilla.com",
                   "tracking-protection.cdn.mozilla.net","location.services.mozilla.com",
                   "detectportal.firefox.com" ,"snippets.cdn.mozilla.net" ]
def open_website(url,count):
    print(count)
    print(url)
    domains = set()
    final_domains = set()
    try :
        driver = webdriver.Firefox()
        driver.set_page_load_timeout(60)
        driver.get(url)
        sleep(2)
        driver.get("about:networking#dns")
        sleep(2)
        table_id =driver.find_element(By.ID, 'dns_content')
        rows = table_id.find_elements(By.TAG_NAME, "tr")
        for row in rows :
            col = row.find_elements(By.TAG_NAME, "td")[0]
            domains.add(col.text)
            print(col.text)
            if not col.text in firefox_domains :
                final_domains.add(col.text)
        print(len(domains))
        print(len(final_domains))
        raw_dns = int(len(domains))
        final_dns = int(len(final_domains))
        site_df.at[count-1 , 'raw_dns'] = raw_dns
        site_df.at[count-1 , 'final_dns'] = final_dns
        sleep(2)
        #driver.close()
    except TimeoutException as ex1:
        print("Exception has been thrown. " + str(ex1))
        driver.get("about:networking#dns")
        sleep(2)
        table_id =driver.find_element(By.ID, 'dns_content')
        rows = table_id.find_elements(By.TAG_NAME, "tr")
        for row in rows :
            col = row.find_elements(By.TAG_NAME, "td")[0]
            domains.add(col.text)
            print(col.text)
            if not col.text in firefox_domains :
                final_domains.add(col.text)
        print(len(domains))
        print(len(final_domains))
        raw_dns = int(len(domains))
        final_dns = int(len(final_domains))
        site_df.at[count-1 , 'raw_dns'] = raw_dns
        site_df.at[count-1 , 'final_dns'] = final_dns
        sleep(2)
        #driver.close()
    except WebDriverException as ex2:
        print("Exception has been thrown. " + str(ex2))
        driver.get("about:networking#dns")
        sleep(2)
        table_id =driver.find_element(By.ID, 'dns_content')
        rows = table_id.find_elements(By.TAG_NAME, "tr")
        for row in rows :
            col = row.find_elements(By.TAG_NAME, "td")[0]
            domains.add(col.text)
            print(col.text)
            if not col.text in firefox_domains :
                final_domains.add(col.text)
        print(len(domains))
        print(len(final_domains))
        raw_dns = int(len(domains))
        final_dns = int(len(final_domains))
        site_df.at[count-1 , 'raw_dns'] = raw_dns
        site_df.at[count-1 , 'final_dns'] = final_dns
        sleep(2)
        #driver.close()
    driver.close()
```

```python
count = 701
for domain in site_df['website'] :
    url = 'http://www.'+domain
    open_website(url,count)
    count = count+1

#open_website("http://www.google.com")
```
